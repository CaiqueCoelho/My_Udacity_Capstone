## Base Estatisticas Medias

```python
base_ita = pd.read_excel("ITALIANO_2020.xlsx")

base_ita = base_ita.drop(["Unnamed: 0","Unnamed: 0.1"],axis=1)

base_ita
```

```python
base_ita_new = base_ita.groupby(['Confronto','Time',"Data"], as_index=False)['Gls','Ast', 'PK', 'PKatt', 'Sh', 'SoT', 'CrdY', 'CrdR',
       'Touches', 'Press', 'Tkl', 'Int', 'Blocks',].sum()



import numpy as np

new_data_1 = []
for i in base_ita_new.Data:
    i = str(i).replace("-","")
    new_data_1.append(i)

base_ita_new["Data_new"] = pd.DataFrame(np.array(new_data_1))




media_sh = []

for i, j in zip(base_ita_new["Data_new"],base_ita_new["Time"]):
    oi = base_ita_new.loc[base_ita_new["Data_new"]<i]
    oi = oi.loc[oi["Time"]==j]
    oi = oi["Sh"].mean()
    media_sh.append(oi)
    
base_ita_new["media_Sh"] = pd.DataFrame(np.array(media_sh))

######

media_sot = []

for i, j in zip(base_ita_new["Data_new"],base_ita_new["Time"]):
    oi = base_ita_new.loc[base_ita_new["Data_new"]<i]
    oi = oi.loc[oi["Time"]==j]
    oi = oi["SoT"].mean()
    media_sot.append(oi)
    
base_ita_new["media_SoT"] = pd.DataFrame(np.array(media_sot))

######

media_crdy = []

for i, j in zip(base_ita_new["Data_new"],base_ita_new["Time"]):
    oi = base_ita_new.loc[base_ita_new["Data_new"]<i]
    oi = oi.loc[oi["Time"]==j]
    oi = oi["CrdY"].mean()
    media_crdy.append(oi)
    
base_ita_new["media_CrdY"] = pd.DataFrame(np.array(media_crdy))

######

media_touches = []

for i, j in zip(base_ita_new["Data_new"],base_ita_new["Time"]):
    oi = base_ita_new.loc[base_ita_new["Data_new"]<i]
    oi = oi.loc[oi["Time"]==j]
    oi = oi["Touches"].mean()
    media_touches.append(oi)
    
base_ita_new["media_Touches"] = pd.DataFrame(np.array(media_touches))

######

media_tkl = []

for i, j in zip(base_ita_new["Data_new"],base_ita_new["Time"]):
    oi = base_ita_new.loc[base_ita_new["Data_new"]<i]
    oi = oi.loc[oi["Time"]==j]
    oi = oi["Tkl"].mean()
    media_tkl.append(oi)
    
base_ita_new["media_Tkl"] = pd.DataFrame(np.array(media_tkl))

######

media_int = []

for i, j in zip(base_ita_new["Data_new"],base_ita_new["Time"]):
    oi = base_ita_new.loc[base_ita_new["Data_new"]<i]
    oi = oi.loc[oi["Time"]==j]
    oi = oi["Int"].mean()
    media_int.append(oi)
    
base_ita_new["media_Int"] = pd.DataFrame(np.array(media_int))

time_fora = []

for i in base_ita_new.Confronto:
    i = str(i)
    i = i.split("_vs_")
    i_1 = str(i[1])
    i_2 = str(i[1])
    a = "Hellas"
    if a in i_1:
        i_1 = i_1.replace("_"," ")
        time_fora.append(i_1)
    else:
        time_fora.append(i_2)
        
        
base_ita_new["Time_Fora"] = pd.DataFrame(np.array(time_fora))


time_casa = []

for i in confrontos.Confronto:
    i = str(i)
    i = i.split("_vs_")
    i_1 = str(i[0])
    i_2 = str(i[0])
    a = "Hellas"
    if a in i_1:
        i_1 = i_1.replace("_"," ")
        time_casa.append(i_1)
    else:
        time_casa.append(i_2)
        
        
base_ita_new["Time_Casa"] = pd.DataFrame(np.array(time_casa))


base_ita_new


```

```python
base_consulta = base_fora_tratada_final.rename(columns={"Time_Fora":"Time"})
base_consulta_final= base_consulta.append(base_dentro_tratada_final)
```

```python
resultado = resultado.replace({"Home": di})
resultado = resultado.replace({"Away": di})

base_nova_casa = pd.DataFrame()

for i in resultado.Home:
    a = i
    procv = base_consulta_final[base_consulta_final.Time.eq(a)]
    sort_procv = procv.sort_values(["Data_new"])
    last_game = sort_procv[len(sort_procv)-1:len(sort_procv)]
    last_game = last_game.reset_index(drop=True)
    last_game = last_game.drop(["Confronto","Data_new",'Gls','sofifa_id'],axis=1)
    base_nova_casa = base_nova_casa.append(last_game)
    
    
base_nova_fora = pd.DataFrame()

for i in resultado.Away:
    a = i
    procv = base_consulta_final[base_consulta_final.Time.eq(a)]
    sort_procv = procv.sort_values(["Data_new"])
    last_game = sort_procv[len(sort_procv)-1:len(sort_procv)]
    last_game = last_game.reset_index(drop=True)
    last_game = last_game.drop(["Confronto","Data_new",'Gls','sofifa_id'],axis=1)
    base_nova_fora = base_nova_fora.append(last_game)

base_nova_casa_op = base_nova_casa[['media_Sh', 'media_SoT', 'media_CrdY', 'media_Touches',
       'media_Tkl', 'media_Int', 'age', 'height_cm', 'weight_kg', 'overall',
       'potential', 'international_reputation', 'weak_foot', 'skill_moves',
       'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']]

base_nova_fora_op = base_nova_fora[['media_Sh', 'media_SoT', 'media_CrdY', 'media_Touches',
       'media_Tkl', 'media_Int', 'age', 'height_cm', 'weight_kg', 'overall',
       'potential', 'international_reputation', 'weak_foot', 'skill_moves',
       'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']]

base_nova_tratada = base_nova_casa_op - base_nova_fora_op
base_nova_tratada = base_nova_tratada.reset_index(drop=True)
base_nova_tratada["Time"] = resultado.Home
base_nova_tratada["Time_Fora"] = resultado.Away
base_nova_tratada["Result"] = ""
base_nova_tratada["N_Gols"] = ""
base_previsao = base_nova_tratada.reset_index(drop=True)
base_previsao = base_previsao[['Time','Time_Fora','media_Sh', 'media_SoT', 'media_CrdY', 'media_Touches', 'media_Tkl',
                               'media_Int', 'age', 'height_cm', 'weight_kg', 'overall', 'potential',
                               'international_reputation', 'weak_foot', 'skill_moves', 'pace','shooting',
                               'passing', 'dribbling', 'defending', 'physic', 'Result', 'N_Gols']]

base_previsao
# teste_bob_2 = pd.concat([teste_bob_2, df4], axis=1, sort=False)
```
